```python
def normalize(df):
    result = df.copy()
    for feature_name in df.columns:
        max_value = df[feature_name].max()
        min_value = df[feature_name].min()
        result[feature_name] = (df[feature_name] - min_value) / (max_value - min_value)
    return result

temp=ex.drop(ex.columns[0:3],axis=1)
temp=normalize(temp)
temp
```

```python
def scale(df):
    result = df.copy()
    for feature_name in df.columns:
        result[feature_name] = df[feature_name]*10
    return result

temp=scale(temp)
temp
```
