```python
def normalize(df):
    result = df.copy()
    for feature_name in df.columns:
        max_value = df[feature_name].max()
        min_value = df[feature_name].min()
        result[feature_name] = ((df[feature_name] - min_value) / (max_value - min_value))*10
    return result

temp=ex.drop(ex.columns[0:2],axis=1)
temp=normalize(temp)
temp
```
